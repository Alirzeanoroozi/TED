```python
import pandas as pd

class Tree_node:
    def __init__(self, name):
        self.name = name
        self.children = {}
        self.leaf = False
        self.count = 0
    
    def add_child(self, name):
        self.children[name] = Tree_node(name)
    
    def get_child(self, name):
        return self.children[name]
    
root_node = Tree_node("root")

def update_tree(parts):
    node = root_node
    for i in range(len(parts)):
        name = ".".join(parts[:i+1])
        if name not in node.children:
            node.add_child(name)
        node = node.get_child(name)
        node.count += 1
    node.leaf = True
    
all_cath_labels = pd.read_csv("patched.ted.csv")['cath_label']
print(all_cath_labels.value_counts())
exit()

cath_labels = all_cath_labels.tolist()
print(len(cath_labels))
for cath_label in cath_labels:
    if cath_label == "-" or pd.isnull(cath_label):
        continue
    update_tree(cath_label.split("."))

# cath_label = "3.90.1150.10"
# parts = cath_label.split(".")
# update_tree(parts)
# cath_label = "3.40.640.10"
# parts = cath_label.split(".")
# update_tree(parts)
# cath_label = "3.40.640.10"
# parts = cath_label.split(".")
# update_tree(parts)

def dfs_print_with_count(root):
    stack = [root]
    while stack:
        node = stack.pop()
        print(node.name, node.count, node.leaf)
        for child in node.children.values():
            stack.append(child)

dfs_print_with_count(root_node)

```
