```python
# Construct the dataset and plot grouped bars for each GPU (AWS-A10, DiffDock)
import matplotlib.pyplot as plt
import pandas as pd


# Raw entries
rows = [
    {"Model": "ESMFold", "GPU": "AWS-A10", "Used_GB": 14.0},
    {"Model": "TMAlign", "GPU": "AWS-A10", "Used_GB": 0.0},
    {"Model": "Diversify", "GPU": "AWS-A10", "Used_GB": 6.7},
    {"Model": "Diffdock", "GPU": "DiffDock", "Used_GB": 6.0},
    {"Model": "Taxonomy", "GPU": "DiffDock", "Used_GB": 3.5},
    {"Model": "Protnlm", "GPU": "DiffDock", "Used_GB": 5.5},
]

df = pd.DataFrame(rows)

# Aggregate by GPU
total_capacity = 23.0
agg = df.groupby("GPU")["Used_GB"].sum().reset_index()
agg["Free_GB"] = (total_capacity - agg["Used_GB"]).clip(lower=0)

# Show table to the user for transparency
# display(agg)

# Prepare data for grouped bar chart
gpus = agg["GPU"].tolist()
used = agg["Used_GB"].tolist()
free = agg["Free_GB"].tolist()

x = range(len(gpus))
width = 0.35

plt.figure(figsize=(8,5))
# Two bars per GPU (Used, Free)
plt.bar([xi - width/2 for xi in x], used, width, label="Used (GB)")
plt.bar([xi + width/2 for xi in x], free, width, label="Free (GB)")

plt.xticks(list(x), gpus)
plt.ylabel("GB")
plt.title("GPU RAM Usage (out of 23 GB per GPU)")
plt.legend()
plt.tight_layout()

# Save a copy
output_path = "/mnt/data/gpu_memory_bars.png"
plt.savefig(output_path, dpi=200)
plt.show()
```
